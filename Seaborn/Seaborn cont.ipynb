## Categorical Plot Types

Categorical data : limited / fixed number of values
maybe compined with numeric data 
Types :<br>
1- abstract representations -> boxplot-violinplot-boxenplot <br>
2-show each observation -> stripplot-swarmplot<br>
3-statistical estimates -> barplot-pointplot-countplot<br>

```python
df = pd.read_csv("https://assets.datacamp.com/production/repositories/2210/datasets/205443d734f177d36dad2f0bdf821a57b2c4cc13/schoolimprovement2010grants.csv")
df
```

```python
# Create the stripplot
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter=True)

plt.show()
```

```python
# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue='Region')

plt.show()
```

```python
# Create a boxplot
sns.boxplot(data=df,
         x='Award_Amount',
         y='Model Selected')

plt.show()
plt.clf()
```

```python
# Create a violinplot with the husl palette
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl')

plt.show()
plt.clf()
```

```python
# Create a boxenplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')

plt.show()
plt.clf()
```

```python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=df,
         y="Model Selected",
         hue="Region")

plt.show()
plt.clf()
```

```python
# Create a pointplot and include the capsize in order to show caps on the error bars
sns.pointplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         capsize=.1)

plt.show()
plt.clf()
```

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue="Region")

plt.show()
plt.clf()
```

## Matrix Plots

To plot a heat map we may use the heatmap() function of seaborn <br>
but it requires an input of a grid format so we use pandas crosstab()<br>
to manipulate the data and make it the right format for seaborn<br>

```python
df_crosstab = pd.crosstab(df["mnth"],df["weekday"],
           values=df["total_rentals"], aggfunc="mean").round(0)
df_crosstab
```

```python
sns.heatmap(pd.crosstab(df["mnth"],df["weekday"],
           values=df["total_rentals"], aggfunc="mean"))
```

```python
sns.heatmap(df_crosstab, annot=True, fmt="", cmap="YlGnBu", cbar=True, linewidths=.5)
```

```python
sns.heatmap(df_crosstab, annot=True, fmt="", cmap="YlGnBu", cbar=True, linewidths=.5, center=df_crosstab.loc[9,6])
```

```python
#plot correlation between variables as heatmap
cols=["total_rentals", "temp", "casual", "hum", "windspeed"]
sns.heatmap(df[cols].corr(), cmap="RdPu_r")
```
