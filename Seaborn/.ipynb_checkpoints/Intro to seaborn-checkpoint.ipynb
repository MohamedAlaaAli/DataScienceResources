we use relplot instead of scatter plot for relational plots

```python
sns.relplot(x="total_bill", y="tip", data=tips,kind="scatter")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips,kind="scatter",col="smoker" )
```

```python
sns.relplot(x="total_bill",
            y="tip", 
            data=tips,
            kind="scatter",
            col="day")
```

### Line Plots

```python
mpg = pd.read_csv("https://assets.datacamp.com/production/repositories/3996/datasets/e0b285b89bdbfbbe8d81123e64727ff150d544e0/mpg.csv")
mpg.head()
```

```python
sns.relplot(data=mpg,
           x="model_year",
           y="mpg",
           kind="line")
#shows the plot of the mean around it 95% confidence interval of the mean 
#not the confidence interval of the distribution
```

```python
sns.relplot(data=mpg,
           x="model_year",
           y="mpg",
           kind="line",
           ci="sd")
#set ci to None to hide any shade
#shows the standard deviation instead of the confidence interval
```

```python
sns.relplot(data=mpg,
           x="model_year",
           y="mpg",
           kind="line",
           ci=None)
```

```python
sns.relplot(x="model_year", 
            y="horsepower", 
            data=mpg, 
            kind="line", 
            ci=None, 
            style="origin", 
            hue="origin")
```

```python
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",
            markers=True,
            dashes=False)
```

## Categorical Plots

```python
survey_data = pd.read_csv("https://assets.datacamp.com/production/repositories/3996/datasets/ab13162732ae9ca1a9a27e2efd3da923ed6a4e7b/young-people-survey-responses.csv")
survey_data.head()
```

### Bar Plot

```python
sns.catplot(data=survey_data,
            x="Gender",
            y="Mathematics",
            kind="bar")
```

```python
student_data = pd.read_csv("https://assets.datacamp.com/production/repositories/3996/datasets/61e08004fef1a1b02b62620e3cd2533834239c90/student-alcohol-consumption.csv")
student_data.head()
```

```python
sns.catplot(x="study_time",
            y="G3",
            data=student_data,
            kind="bar")
```

```python
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Rearrange the categories
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar", 
            order=category_order,
           ci=None)
```

### Box Plot

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(data=student_data,
            x="study_time",
            y="G3",
            kind="box",
            order=study_time_order)
```

```python
sns.catplot(data=student_data,
            x="study_time",
            y="G3",
            kind="box",
            order=study_time_order,
           sym="")
#sym can omit the outliers
```

```python
sns.catplot(data=student_data,
            x="internet",
            y="G3",
            kind="box",
            hue="location",sym="")
```

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=0.5)
#whis chgange the equation of the whiskers to make it 0.5*IQR instead of 1.5*IQR
#whis can take also a list of percentiles or max and min bounds
```

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5,95])
#5th to 95th percentiles
```

```python
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0,100])
# zero and 100% of percentiles 
```

### Point Plots

shows the mean of quantitative variable 
vertical line represent 95% confidence
on condition that data is random samples
one axis is categorical while another is quantitative

```python
sns.catplot(x="smoker",
           y="total_bill",
           data=tips,
           kind="point")
```

```python
# change for the median instead of the mean due to means sensitivity to outliers
import numpy
sns.catplot(x="smoker",
           y="total_bill",
           data=tips,
           estimator=numpy.median,
           kind="point",
           capsize=0.2)
#capsize adds caps to the end of confidence intervals 
```

```python
sns.catplot(x="smoker",
           y="total_bill",
           data=tips,
           kind="point",
           ci=None)
```

```python
sns.catplot(x="smoker",
           y="total_bill",
           data=tips,
           kind="point",
           capsize=0.2,
           join=False)
```
